```python
#opening web driver
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(5)

#giving url to webdriver
url='https://www.google.co.in/maps/'
driver.get(url)
time.sleep(5)
#searching xpath on webpage
search_bar=driver.find_element_by_xpath('//input[@class="tactile-searchbox-input"]')
time.sleep(3)

#giving input in search bar
search_bar.send_keys(" new delhi")
time.sleep(3)

#searching data and pressing button
search_button=driver.find_element_by_xpath('//button[@class="searchbox-searchbutton"]')
search_button.click()
time.sleep(10)
```

```python
#extracting url for scraping
try:
    url_string = driver.current_url   #extracting current web url
    print("URL Extracted: ", url_string)  #print url

    lat_lng = re.findall(r'@(.*)data',url_string)

    if len(lat_lng): #extracting lan of latitude and longitude

        lat_lng_list = lat_lng[0].split(",") #sliting the url

        if len(lat_lng_list)>=2:

            lat = lat_lng_list[0] #print latitude

            lng = lat_lng_list[1] #print longitude

        print("Latitude = {}, Longitude = {}".format(lat, lng))



except Exception as e:  # if not iterable

        print("Error: ", str(e))
```

```python
#opening web driver
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(3)

#giving url to webdriver
url='https://www.hostelworld.com/'
driver.get(url)
time.sleep(5)

#searching xpath on webpage
search_bar=driver.find_element_by_xpath('//input[@class="location-text"]')

#giving input in search bar
search_bar.send_keys("London, England")

#searching data and pressing button
search_button=driver.find_element_by_xpath('//button[@class="button primary large"]')
search_button.click()
```

```python
#assigning empty list for data
name_of_hostal=[]
distance_from_citycenter=[]
total_rating=[]
overall_review=[]
reviews=[]
private_price=[]
privare_room_price=[]
dorms_room_price=[]
room_facilities=[]
hostel_description=[]

#fetching xpath for the hostel data
name=driver.find_elements_by_xpath("//h2[@class='title title-6']")
distance=driver.find_elements_by_xpath("//span[@class='description']")
rating=driver.find_elements_by_xpath("//div[@class='score orange big']")
review1=driver.find_elements_by_xpath("//div[@class='keyword']")
review=driver.find_elements_by_xpath("//div[@class='reviews']")
price=driver.find_elements_by_xpath("//div[@class='price title-5']")
facilty=driver.find_elements_by_xpath("//div[@class='facilities-label facilities']")
description=driver.find_elements_by_xpath("//div[@class='rating-factors prop-card-tablet rating-factors small']")
```

```python
#fetching data for name of hostel
for i in name:
    name_of_hostal.append(i.text)

#fetching data for distance of hostel from city center    
for i in distance:
    distance_from_citycenter.append(i.text.replace('Hostel - ',''))

#fetching data for rating of the hostel    
for i in rating:
    total_rating.append(i.text)

#fetching data for review of hostel    
for i in review1:
    overall_review.append(i.text)

#fetching data for overall review of hostel
for i in review:
    reviews.append(i.text)

#fetching data for prices of rooms of hostel
for i in price:
    private_price.append(i.text)
    
#fetching data for private room price of hostel
for i in range(0,len(private_price),2):
    privare_room_price.append(private_price[i])
    
#fetching data for dorms room price of hostel   
for i in range(1,len(private_price),2):
    dorms_room_price.append(private_price[i].replace('\n',''))

#fetching data for faciliteis offered by hostel
for i in facilty:
    room_facilities.append(i.text)

#fetching data for discription of the hostel
for i in description:
    try:
        hostel_description.append(i.text.replace('\n',''))
        
    except:
         hostel_description.append('-')
```

```python
hostels= pd.DataFrame({})
hostels['name']= name_of_hostal[:20]
hostels['distance from citycenter']= distance_from_citycenter[:20]
hostels['rating']= total_rating[:20]
hostels['overall review']= overall_review[:20]
hostels['privates from price']= privare_room_price[:20]
hostels['dorms from price']= dorms_room_price[:20]
hostels['facilities']= room_facilities[:20]
hostels['description']= hostel_description[:20]
hostels
```
