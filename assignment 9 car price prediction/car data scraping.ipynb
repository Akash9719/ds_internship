```python
# Importing Libraries
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup

# Importing selenium webdriver 
from selenium import webdriver

# Importing required Exceptions which needs to handled
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

#Importing requests
import requests

# importing regex
import re
```

```python
# Activating the chrome browser
driver = webdriver.Chrome("chromedriver.exe") 
time.sleep(3)

# Opening the homepage of carwale
url = "https://www.cartrade.com/buy-used-cars/#sc=-1&so=-1&pn=1"
driver.get(url)
time.sleep(3)
```

```python
car_url=[]
#scrapping the required details
start=0
end=100
for page in range(start,end):#for loop for scrapping 4 page
    url =driver.find_elements_by_xpath("//h2[@class='h2heading truncate']//a")
    for i in url:
        car_url.append(i.get_attribute("href"))
    nxt_button=driver.find_elements_by_xpath("//li[@class='next']//a")
    try:
        driver.get(nxt_button[1].get_attribute('href'))#getting the link from the list for next page
    except:
        driver.get(nxt_button[0].get_attribute('href'))
        
```

```python
df=pd.DataFrame({"car_url":car_url})
df
```

```python
car_url
```

```python
car_brand=[]
car_name=[]
manufacturing_year=[]
location=[]
fuel_type=[]
kilometers_driven=[]
number_of_owners=[]
price=[]

for i in car_url:
    driver.get(i)
        
    try:
        brand = driver.find_element_by_xpath("//div[@class='wrapper v_content']//ul//li[4]")
        car_brand.append(brand.text)
    except:
        car_brand.append('-')
        
    try:
        name = driver.find_element_by_xpath("//div[@class='wrapper v_content']//ul//li[5]")
        car_name.append(name.text)
    except:
        car_name.append('-')
        
    try:
        year = driver.find_element_by_xpath("//table[@class='v_table']//tr[8]//td[2]")
        manufacturing_year.append(year.text)
    except:
        manufacturing_year.append('-')
        
    try:
        locations = driver.find_element_by_xpath("//table[@class='v_table']//tr[1]//td[2]")
        location.append(locations.text)
    except:
        location.append('-')
        
    try:
        fuel = driver.find_element_by_xpath("//table[@class='v_table']//tr[2]//td[2]")
        fuel_type.append(fuel.text)
    except:
        fuel_type.append('-')

    try:
        kms = driver.find_element_by_xpath("//table[@class='v_table']//tr[3]//td[2]")
        kilometers_driven.append(kms.text)
    except:
        kilometers_driven.append('-')
        
    try:
        owners = driver.find_element_by_xpath("//table[@class='v_table']//tr[5]//td[2]")
        number_of_owners.append(owners.text)
    except:
        number_of_owners.append('-')
        
    try:
        car_price = driver.find_element_by_xpath("//div[@class='pull-left']//span[2]")
        price.append(car_price.text)
    except:
        price.append('-')
```

```python
df=pd.DataFrame({"car_brand":car_brand,"car_name":car_name,
                "manufacturing_year":manufacturing_year,"location":location,"fuel_type":fuel_type,"kilometers_driven":kilometers_driven,"number_of_owners":number_of_owners,"price":price})
df
```

```python
df.to_excel('df.xlsx')
```
