 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#opening web driver
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(2)

#giving url to webdriver
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
```

```python
#Assigning empty list
ranks=[]
name_of_video=[]
artists=[]
upload_date=[]
views_of_video=[]
time.sleep(3)

#scraping the xpath for finding the data regard empty lists
table_data=[]
data=driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tr//td")
for i in data:
    if i.text is None :
        table_data.append("-") 
    else:
        table_data.append(i.text)

#fetching the rank of video from table data
rank_of_video=[]     
for i in range(0,len(table_data),6):
    rank_of_video.append(table_data[i])
    time.sleep(4)

#fetching the name of video from table data
for i in range(1,len(table_data),6):         
    name_of_video.append(table_data[i])
    time.sleep(4)
    
#fetching the artist of video from table data
for i in range(2,len(table_data),6):     
    artists.append(table_data[i])
    time.sleep(4)
    
#fetching the upload date of video from table data
for i in range(4,len(table_data),6):     
    upload_date.append(table_data[i])
    time.sleep(4)

#fetching the views of video from table data
for i in range(3,len(table_data),6):      
    views_of_video.append(table_data[i])
```

```python
#making dataframe of above data
video = pd.DataFrame({})     
video['rank']=rank_of_video[:30]
video['name']= name_of_video[:30]
video['artist']= artists[:30]
video['upload date']= upload_date[:30]
video['views']= views_of_video[:30]
video
```

```python
#assigning empty lists
dataset_name=[]
data=[]
data_types=[]
attribute_type=[]
no_of_instance=[]
attributes=[]
year=[]


#scraping the repository titles
dataset_name=[]
name=driver.find_elements_by_xpath("//p[@class='normal']//b//a")
for i in name:
    if i.text is None :
        dataset_name.append("-") 
    else:
        dataset_name.append(i.text)
        
#scraping the repository other data
data_s=driver.find_elements_by_xpath("//p[@class='normal']")
for i in data_s:
    if i.text is None :
        data.append("-") 
    else:
        data.append(i.text)
        
#fetching data types of repositery through xpath
for i in range(9,len(data),7):
    data_types.append(data[i])
    
#fetching tasks of repositery through xpath    
task=[]
for i in range(10,len(data),7):
    task.append(data[i])
    
#fetching attribute type of repositery through xpath     
for i in range(11,len(data),7):
    attribute_type.append(data[i])
    
#fetching number of instance of repositery through xpath      
for i in range(12,len(data),7):
    no_of_instance.append(data[i])
    
#fetching attributes lists of repositery through xpath    
for i in range(13,len(data),7):
    attributes.append(data[i])
    
#fetching year of uploading of repositery through xpath    
for i in range(14,len(data),7):
    year.append(data[i])
    
#meking data frame from above data    
ML=pd.DataFrame({})
ML['dataset name'] = dataset_name[:100]
ML['data type'] = data_types[:100]
ML['task'] = task[:100]
ML['attribute type'] = attribute_type[:100]
ML['no of instance'] = no_of_instance[:100]
ML['no of attributes'] = attributes[:100]
ML['year'] = year[:100]
ML
```
